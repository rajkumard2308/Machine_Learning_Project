## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report, roc_auc_score,roc_curve
from imblearn.over_sampling import SMOTE
```

## Loading Dataset

```python
data = pd.read_csv("creditcard.csv")
```

## Data Cleaning and Preprossening

```python
# Check Null Values
data.isnull().sum()
```

```python
# Check Duplicated Values
data.duplicated().value_counts()
```

```python
# Drop Duplicated Values from Dataset
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

## Feature Selection

```python
# Drop unnecessary features
data.drop(columns=['Time'],inplace=True)
data.head()
```

## Feature Scaling

```python
scaler = StandardScaler()
data[['Amount']] = scaler.fit_transform(data[['Amount']])
```

```python
data.head()
```

## Spilit into input and output feature

```python
X = data.drop(['Class'],axis=1)
y = data['Class']
```

## Use Over Sampling Metohd(SMOTE) over Imbalenced Dataset

```python
smote = SMOTE(random_state=42)
X_smote,y_smote = smote.fit_resample(X,y)

print("Before SMOTE:", y.value_counts())
print("After SMOTE:", y_smote.value_counts())
```

## Train Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X_smote,y_smote,test_size=0.2,random_state=42)
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
```

## Train Model with Various Algoritms

## 1. Logistic regression

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred1=lr.predict(X_test)
y_prob1 = lr.predict_proba(X_test)[:,1]
```

```python
print(f"Accuracy score : {accuracy_score(y_test,y_pred1)}")
print(f"Precision score : {precision_score(y_test,y_pred1)}")
print(f"recall score : {recall_score(y_test,y_pred1)}")
print(f"F1 score : {f1_score(y_test,y_pred1)}")
print(classification_report(y_test, y_pred1))
print("ROC-AUC:", roc_auc_score(y_test, y_prob1))
```

## Plot ROC Curve

```python
# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob1)

# Compute AUC
auc_score = roc_auc_score(y_test, y_prob1)

# Plot
plt.figure()
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {auc_score:.4f})")
plt.plot([0, 1], [0, 1], linestyle='--')  # random model
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.title("ROC Curve - Credit Card Fraud Detection")
plt.legend()
plt.show()
```

## 2. DecisionTreeClassifier

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
print(f"Accuracy score : {accuracy_score(y_test,y_pred2)}")
print(f"Precision score : {precision_score(y_test,y_pred2)}")
print(f"recall score : {recall_score(y_test,y_pred2)}")
print(f"F1 score : {f1_score(y_test,y_pred2)}")
print(classification_report(y_test, y_pred2))
print("ROC-AUC:", roc_auc_score(y_test, y_prob2))
```

```python
FEATURE_ORDER = [
    "V1","V2","V3","V4","V5","V6","V7","V8","V9","V10",
    "V11","V12","V13","V14","V15","V16","V17","V18","V19","V20",
    "V21","V22","V23","V24","V25","V26","V27","V28","Amount"
]
# Reorder input dataframe
input_df = input_df[FEATURE_ORDER]
```

## Make Prediction

```python
# Predict probability
fraud_prob = model.predict_proba(input_df)[0][1]

# Apply threshold
threshold = 0.3
prediction = "FRAUD" if fraud_prob > threshold else "LEGIT"

print(f"Prediction: {prediction}")
print(f"Fraud Probability: {fraud_prob:.4f}")
```
